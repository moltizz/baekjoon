```python
H, M = map(int, input().split())

if H >= 0 and H <= 23 and M >= 0 and M <= 59:
    if H > 0 and M < 45:
        print(H-1, M+15)
    elif H > 0 and M == 45:
        print(H, 0)
    elif H > 0 and M > 45:
        print(H , M-45)
    elif H == 0 and  M < 45:
        print(H+23,M+15)
    elif H == 0 and M == 45:
        print(0, M-45)
    elif H == 0 and M > 45:
        print(0, M-45)

# 다섯 번만에 맞혔다.
```

```python
# 첫 번째 시도
H, M = map(int, input().split())

if H >= 0 and H <= 23 and M >= 0 and M <= 59:
    if H > 0:
        print(H-1, M+15)
    else:
        print(H+23, M+15)

# 너무 깊게 생각하지 않았다.
```

```python
# 두 번째 시도
H, M = map(int, input().split())

if H >= 0 and H <= 23 and M >= 0 and M <= 59:
    if H > 0 and M < 45:
        print(H-1, M+15)
    elif H >= 0 and M == 45:  # <1> 여기서 틀림
        print(H+23, 0)
    elif H > 0 and M > 45:
        print(H , M-45)
    elif H == 0 and  M < 45:
        print(H+23,M+15)
    elif H == 0 and M > 45:
        print(0, M-45)

# <!> H = 0을 합쳐도 된다는 단순한 사고를 했다.
```

```python
# 세 번째 시도
H, M = map(int, input().split())

if H >= 0 and H <= 23 and M >= 0 and M <= 59:
    if H > 0 and M < 45:
        print(H-1, M+15)
    elif H > 0 and M == 45:
        print(H+23, 0)  # <1> 여기서 틀림
    elif H > 0 and M > 45:
        print(H , M-45)
    elif H == 0 and  M < 45:
        print(H+23,M+15)
    elif H == 0 and M == 45:
        print(0, M-45)
    elif H == 0 and M > 45:
        print(0, M-45)

# <!> elif 구문이 늘어나다 보니까 기본적으로 맞아야 하는 print 인자를 잘못 생각했었다.
```
