# A+B-4

입력이 끝날 때까지 A+B를 출력하는 문제. EOF에 대해 알아 보세요.

image.png

### 성공

```python
while True:
  try:
      a,b = map(int,input().split())
      print(a+b)
  except:
      break
```

**✍ 회고**

문제에서 EOF에 대해 알아보라는 힌트를 주었다. 그래서 구글링을 하되, 해당 문제에 대한 설명이 직접적으로 드러나지 않는 글을 찾으려고 했다. 관련된 예제 위주로 찾아봤는데, EOF 자체의 틀이 있어서 여러 글들의 내용이 일관적이었다. 그래서 쉽게 성공할 수 있었다.

문자열을 입력받아 EOF(End Of File;파일의 끝)를 판단할 수 있는지를 묻는다. 입력으로 테스트 케이스의 개수가 주어지지 않으므로 EOF를 판단해서 프로그램을 종료해야 한다. 해당 문제에 적용한 EOF를 판단법은 반복문 안에 try, except를 사용해서 EOFError를 처리한다.
