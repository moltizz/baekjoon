###오답

```python
# 실패
T = int(input())
a,b = list(map(int, input().split()))

T <= 1000000 and T >= 0

for i in range(T):
  print(a+b)
```

for문에 대한 제대로 된 개념이 적립되지 않았다. for문의 개념을 구글링해도 이해하는 것이 쉽지 않다. 그래서 정답을 보고 이해하기로 하였다.

###구글링 정답

```python
import sys
t = int(input())

for _ in range(t):
  a, b =map(int,input().split())
  print(a+b)
```

**✍ 회고**

for문은 많이 사용되는 구문이다. 구글링을 해서 나오는 for문의 개념도 잘 이해하지 못 하였다. 그래서 정답을 보고 이해하고자 하였다. 정답을 보니까 나의 수준이 심각하다는 것을 깨달았다... 오답이 된 가장 큰 원인은 이전 단계인 '입출력과 사칙연산'과 '조건문'에서 맨 윗줄에 변수=input() 형식으로 입력을 받아왔고, 이런 부분이 입력을 받을 때 맨 윗줄에 다 있어야 한다는 인식이 박혔다. 너무 정확한 오답이었어서 아쉽진 않은데, for문에 대한 방향성은 바로 잡았다고 본다.

추가로 문제에서 T는 최대 1,000,000이라고 해서 따로 범위를 설정해줘야 한다고 생각했는데, 정답을 보니까 그럴 필요가 없어 보인다. 센스있게 문제만 맞게 풀면 되는 것인가 보다.
