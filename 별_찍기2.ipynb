###성공

```python
t = int(input())

for i in range(t):
  i += 1
  x = '*'*i
  print(x.rjust(t))
```

구글링을 통해 rjust를 사용해야 하는 것을 깨달았고 성공하였다.

### 실패

```python
t = int(input())

for i in range(t):
  i += 1
  
  print(' '*(t-i),'*'*i)
```

오로지 내 생각을 통해 도출한 방법이다. 출력을 해보면서 나름 오른쪽 정렬을 성공하였다. 하지만 맨 왼쪽 첫칸에 공백이 생겼다. 그래서 제출 결과, '출력 형식이 잘못되었습니다.'라는 문구가 떴다.

**✍ 회고**

처음에는 어떻게 오른쪽 정렬을 해야 할지 몰라서 막막했다. 구글링을 해도, 구글링에 나온 걸 여기에 적용시켜도 되는지, 적용시킬 수 있는지 등 잘 몰랐다. 내가 아는 선에서 시도해볼 방법이 input으로 출력되는 별의 나머지 수를 여백으로 채우는 것이었다. 오른쪽 정렬엔 성공하였으나, 제대로 된 오른쪽 정렬이 아니라 실패하였다. 다시 구글링을 하여 '파이썬 오른쪽 정렬' 관련 글들을 읽었더니 rjust를 시도해보았다. 결과는 성공적이었다. 약 한달 반 전에 이 문제를 봤을 땐 절대 풀 수 없었는데, 지금 풀 수 있다는 사실이 뿌듯해진다.
