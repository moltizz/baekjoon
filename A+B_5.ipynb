### 성공

```python
while True:
    a, b = map(int, input().split())
    if a == 0 and b == 0:
        break
    print(a+b)
```

계속해서 실패가 떴는데, 마지막에 if문과 print의 위치를 바꾸었더니 성공하였다.

### 실패

```python
# 실패 유형1
for _ in range(_):
    a, b = map(int, input().split())
    print(a+b)
    
    if a == 0 and b == 0:
        break
```

이전까지 for문을 사용해서 해당 문제도 for문일 것이라 생각했다. 그리고 0 0을 넣었을 때 반복문이 멈춰야 하므로 break를 넣은 것이다. for문에 break까지 넣고... 코린이 인증이다. 

```python
# 실패 유형2
a > 0 and b < 10

while True:
    a, b = map(int, input().split())
    print(a+b)
    
    if a == 0 and b == 0:
        break
```

for문에서 while문으로 갈피를 제대로 잡았고, break까지 완벽하게 써줬다. 하지만 0 0을 넣었을 때, 반복문에서 빠져나와야 하는데, 0 0을 더한 값인 0까지 출력되었다. 구글링을 해서 마지막 출력을 없애는 방법을 찾아보았다. 당연히 안나왔다. 고민을 더 해보니까 이 문제는 구문의 순서만 바꿔주면 해결되는 문제였다.

```python
# 실패 유형3
a > 0 and b < 10

while True:
    a, b = map(int, input().split())
    if a == 0 and b == 0:
        break
    print(a+b)
```

위와 같이 제출했을 때 '런타임 에러'가 떴다. 입출력도 정상적으로 작동하는데 왜 뜨는지 알 수 없었다. 맨 윗줄에 적은 a와 b의 대소관계가 마음에 걸려서 지워봤다. 그리고 제출을 했는데 성공했다. 문제에서 (0<A, B<10)이라는 내용이 있어서 적어준 조건인데, 왜 적으면 안되는지 잘 모르겠다. 이전 문제에서 이런 유형의 문제가 있었던 걸로 기억한다.

**✍ 회고**

이번 문제는 매우 쉬울 것이라 생각했다. 그런데 생각보다 술술 풀리진 않았다. 반복문이라고 해서 for문만 쓰려고 했던 점, 코드의 순서에 따라 출력이 바뀐다는 점 등 놓치고 있던 부분이 많았다는 걸 느꼈다. 하지만 이전보다 스스로 생각하면서 결과를 도출해내고 있다. 백준을 시작하기 전, 반복문이 제일 어렵게만 느껴졌는데 하나씩 성공하니 마음이 편안해진다.
